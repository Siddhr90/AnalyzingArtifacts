```python
import pandas as pd
df_a2 = pd.read_json('eval_predictions_SNLI_updated_with_ANLI.jsonl', lines=True)
df_a2.head()
```

```python
import pandas as pd

# Assuming df_base and df_a1 are your DataFrames

# Step 1: Merge the DataFrames on 'premise' and 'hypothesis'
merged_df = pd.merge(df_base, df_a2, on=['premise', 'hypothesis'], suffixes=('_base', '_a1'))

# Step 2: Filter rows where label matches predicted_label in df_base but not in df_a1
filtered_df = merged_df[
    (merged_df['label_base'] != merged_df['predicted_label_base']) &
    (merged_df['label_a1'] == merged_df['predicted_label_a1'])
]

# Step 3: Selecting relevant columns (optional)
# You can modify this list to include the columns you're interested in
columns_to_display = ['premise', 'hypothesis', 'label_base', 'predicted_label_base', 'label_a1', 'predicted_label_a1']
result_df = filtered_df[columns_to_display]

# Display the result
result_df
```
